# Harmonic oscilator : $\frac{1}{2}m\omega^2q^2$

```python
import numpy as np
import math as m
from pylab import plot, xlabel, ylabel, show, legend

m=0.01 #질량이 0.01인 입자를 생각하자.

def V(q):
    omega=1.0
    pot=0.5*m*omega**2*q**2
    return pot

def T(p):
    return 0.5*p**2/m

def H(q,p):
    return T(p)+V(q)
```

해밀토니안이 시간에 대한 상수라고 하고 $q_0 $을 -1이라 하자. 

Hamilton eqaution에서 phase space위의 점에 대한 움직임을 알아낼 수 있다.

$$-\frac{\partial H}{\partial q}=\dot{p}, \quad \frac{\partial H}{\partial p}=\dot{q}$$

1-dim harmonic oscilator의 경우에 쉽게 계산할 수 있다.

$$\dot{p}=-m\omega^2q,\quad \dot{q}=\frac{p}{m}$$

두 식을 연립하면 2차 미방 두 개가 나온다.

$$\ddot{q}+\omega^2q=0,\quad \ddot{p}+\omega^2p=0$$ 

지금 상황을 정리하면,

$$q_0=-1, \quad p_0=0, \quad H_0=H(-1,0)=const. $$

```python
def fq(r,t):   # q''+ w^2 q = 0 미방    v = q'
    q=r[0]
    v=r[1]
    f_q=v
    f_v=-(omega**2)*q
    return array([f_q,f_v],float)

#init. condit. 
t_i=0.0
q_i=-1.0
v_i=0.0

t_f=5.0
h=0.0001
pt=[]
pq=[]
pv=[]
pp=[]
r=[q_i, v_i]

q=q_i
v=v_i
t=t_i

while t<=t_f:
    k1=h*f(r,t)
    k2=h*f(r+0.5*k1, t+0.5*h)
    k3=h*f(r+0.5*k2, t+0.5*h)
    k4=h*f(r+k3, t+h)
    r+=(k1+2*k2+2*k3+k4)/6
    t+=h
    pq.append(r[0])
    pv.append(r[1])
    pp.append(m*r[1]) # 이렇게 간단한 경우에 p=mq'이다.
    pt.append(t)
```

```python
plot(pt,pq,'r')  #red
plot(pt,pv,'b')  # blue
xlabel(r'$t$', fontsize=20)
ylabel(r'$q$, $\dot{q}$', fontsize=20)
show()
```

```python
plot(pq, pv)
xlabel(r'$q$', fontsize=20)
ylabel(r'$\dot{q}$', fontsize=20)
show()
```

이렇게 간단한 경우,  $p=m\dot{q}$

```python
plot(pq, pp)
xlabel(r'$q$', fontsize=20)
ylabel(r'$p$', fontsize=20)
show()
```

```python
pV=[]
pH=[]

for qq in pq:
    pV.append(V(qq))
    pH.append(H(-1,0))
    
plot(pq,pV,'r')
plot(pq,pH,'g')

show()
```

```python
def fq(r,t):   # q''+ w^2 q = 0 미방    v = q'
    q=r[0]
    v=r[1]
    f_q=v
    f_v=-(omega**2)*q
    return array([f_q,f_v],float)

#init. condit. 
t_i=0.0
q1_i=-1.0
q2_i=-0.9
q3_i=-0.3
v1_i=0.0
v2_i=0.0
v3_i=0.0

t_f=5.0
h=0.0001
pt=[]
pq1=[]
pv1=[]
pq2=[]
pv2=[]
pq3=[]
pv3=[]
pp1=[]
pp2=[]
pp3=[]
r1=[q1_i, v1_i]
r2=[q2_i, v2_i]
r3=[q3_i, v3_i]

q1=q1_i
v1=v1_i
q2=q2_i
v2=v2_i
q3=q3_i
v3=v3_i
t=t_i

while t<=t_f:
    k11=h*f(r1,t)
    k21=h*f(r1+0.5*k11, t+0.5*h)
    k31=h*f(r1+0.5*k21, t+0.5*h)
    k41=h*f(r1+k31, t+h)
    r1+=(k11+2*k21+2*k31+k41)/6
    
    k12=h*f(r2,t)
    k22=h*f(r2+0.5*k12, t+0.5*h)
    k32=h*f(r2+0.5*k22, t+0.5*h)
    k42=h*f(r2+k32, t+h)
    r2+=(k12+2*k22+2*k32+k42)/6
    
    k13=h*f(r3,t)
    k23=h*f(r3+0.5*k13, t+0.5*h)
    k33=h*f(r3+0.5*k23, t+0.5*h)
    k43=h*f(r3+k33, t+h)
    r3+=(k13+2*k23+2*k33+k43)/6
   
    t+=h
    pq1.append(r1[0])
    pv1.append(r1[1])
    pp1.append(m*r1[1]) # 이렇게 간단한 경우에 p=mq'이다.
    
    pq2.append(r2[0])
    pv2.append(r2[1])
    pp2.append(m*r2[1])
    
    pq3.append(r3[0])
    pv3.append(r3[1])
    pp3.append(m*r3[1])
    pt.append(t)
```

```python
pV1=[]
pH1=[]

pV2=[]
pH2=[]

pV3=[]
pH3=[]

for qq1 in pq1:
    pV1.append(V(qq1))
    pH1.append(H(q1_i,0))
    
for qq2 in pq2:
    pV2.append(V(qq2))
    pH2.append(H(q2_i,0))
    
for qq3 in pq3:
    pV3.append(V(qq3))
    pH3.append(H(q3_i,0))
```

```python
plot(pq1, pp1,'b')
plot(pq2, pp2,'b')
plot(pq3, pp3,'b')
xlabel(r'$q$', fontsize=20)
ylabel(r'$p$', fontsize=20)
show()
```

```python
plot(pq1,pV1,'r')
plot(pq1,pH1,'g')

plot(pq2,pV2,'r')
plot(pq2,pH2,'g')

plot(pq3,pV3,'r')
plot(pq3,pH3,'g')

show()
```
