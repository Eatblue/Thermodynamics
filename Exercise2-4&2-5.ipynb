# 연습문제 2-4

질량이 $m$ 인 자유입자가 길이가 $L$인 1차원 상자에 갇혀 있을 때, 계의 가능한 에너지가 다음과 같이 주어진다.

$$E=\frac{h^{2}}{8mL^{2}}n^{2}, \quad n=1,2,3,\cdots.$$

(가) 에너지가 $E$보다 작은 계의 상태수를 구해라

$$ \Omega(E)=n=\sqrt{\frac{8mL^{2}}{h^{2}}}E^{1/2}$$

(나) 에너지가 $E \sim E+\delta E$ 사이에 있는 계의 상태수를 구하여라

에너지가 $E \sim E+\delta E$ 사이에 있는 상태수를 $\Omega(E,\delta)$ 라고 하자. 그러면 다음 식이 성립한다. (식 2-14)

$$ \Omega(E,\delta)=\Omega(E+\delta)-\Omega(E)=\frac{\partial\Omega(E)}{\partial E}\delta E$$

(가)에서  $ \Omega(E)=n=\sqrt{\frac{8mL^{2}}{h^{2}}}E^{1/2}$ 이므로 

$$\frac{\partial \Omega}{\partial E}=\sqrt{\frac{2mL^{2}}{h^{2}}}E^{-1/2}$$

따라서, 

$$ \Omega(E,\delta)=\sqrt{\frac{2mL^{2}}{h^{2}}}E^{-1/2}\delta E$$

(다) 이 입자가 각 변의 길이가 $L_{x}=L_{y}=L_{z}=L$인 3차원 상자 속에 있을 때, 에너지가 $E \sim E+\delta E$ 사이에 있는 계의 상태수를 구하여라

총 에너지는 모든 방향 에너지의 합이다; 

$$E=\frac{h^{2}}{8mL^{2}}({n_x}^{2}+{n_y}^{2}+{n_z}^{2})\qquad n_{x},n_{y},n_{z}=1,2,3,\cdots.$$

이때 $n^{2}={n_x}^{2}+{n_y}^{2}+{n_z}^{2}$라고 하면, $({n_x}^{2},{n_y}^{2},{n_z}^{2})$ 순서쌍은 반지름이 $n$인 1/8구 (중심이 원점인 구의 1사분면 부분) 내부에 있는 격자들의 집합과 같다.

(왜냐하면 $n_{x},n_{y},n_{z}$ 모두 양의 정수만 올 수 있기 때문에)

![alt text](images/nsquare.png "nsqaure")

따라서 위 그림을 식으로 표현하면 

$$\frac{1}{8}\left( \frac{4\pi}{3} n^{3} \right),\quad E=\frac{h^{2}}{8mL^{2}}(n^{2})$$

즉

$$\Omega(E)=\frac{\pi L^{3}}{6\,h^{3}}(8mE)^{3/2}$$

에너지가 $E \sim E+\delta E$ 사이에 있을 계의 상태수 $\Omega(E,\delta E)$는

$$\Omega(E,\delta E)=\frac{\partial \Omega}{\partial E}\delta E=\frac{4 \pi m L^{3}}{h^3}(2mE)^{1/2}\delta E$$ 이다.

# 연습문제 2-5

연습무제 (2-4)와 같은 자유입자가 3개 있다. 모든 가능한 에너지상태의 에너지 값과 상태수를 구하여라. 단 에너지 값은 $8mL^{2}/h^{2}$를 단위로 하여 차원이 없는 숫자로 표시해라. 

길이 $L$인 1차원 상자에 세 입자가 있다고 가정하자. 이것은 가로,세로,높이가 각각 길이 $L$인 3차원 상자에 한 입자가 있는 상태와 같다.

입자 x,y,z의 에너지를 각각 ${n_x}^2,{n_y}^2,{n_z}^2$ 라고 하자. 가능한 에너지 상태들은 $({n_x},{n_y},{n_z})$의 순서쌍으로 나타낼 수 있다. 이 순서쌍들을 pandas 패키지를 이용해서 표로 나타내보자.

```python
import numpy as np
from pandas import Series, DataFrame
```

```python
nmat=np.array([[0,0,0,0,0]])#더미 세트/ placeholder가 있으면 더미셋을 안써도 됌..

for i in range(5):  # (nx ny nz)순서쌍 생성 루프
    x=i+1
    for j in range(i+1):
        y=j+1
        for k in range(j+1):
            z=k+1
            if i==k:    #상태수를 정해주기 위해서 만든 조건문 i 가 k와 같다면  i j k 모두 같은 경우임! -> 1가지
                state=1
            elif i!=j and j!=k:  #모두 다를 경우의 수 -> 3 ! 가지
                state=6 
            else:
                state=3 # 나머지 경우 3 C 1 or 3 C 2
            totE=x**2+y**2+z**2  #전체 에너지
            nmat=np.append(nmat,[[x,y,z,totE,state]],axis=0)   #더미세트에 합쳐주기
nmat
```

pandas의 데이터프레임을 이용해서 깔끔하게 표로 정리하자.

```python
frame=DataFrame(nmat,columns=['nx','ny','nz','totE','state'])
frame
```

에너지의 변화를 그림으로 나타내보자. 

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(frame.totE)
plt.show()
```

이온화에너지 경향성처럼 값이 꺾이는 부분이 종종 나타난다. 이럴 경우에 에너지 크기대로 정렬해주고 싶다면 pandas패키지의 sort_values 함수를 쓰면 된다.

```python
frame=frame.sort_values(by=['totE'])
frame
```

totE 항목이 정렬된 것이 보인다. 그림을 그려보면

```python
plt.plot(frame.totE)
plt.show()
```

앞의 인덱스가 정리가 안돼서 이리저리 꼬인 모습을 보인다. 인덱스를 재정렬하기 위해 pandas 패키지의 reset_index 함수를 쓰자

```python
frame=frame.reset_index(drop=True)
frame
```

```python
'''
모델들
'''
def arcHyptan(x,a,b):
    y=0.5*(np.log(1+(a*x+b))-np.log(1-(a*x+b)))   # abs(a*x+b)<1이어야 함!
    return y

def func1(x,a,b,c,d):
    y=c*(arcHyptan(x,a,b)+d)
    return y

def func2(t, a, b):
    return a + b * np.log(t)

def func3(x,a,b):
    return a*arcHyptan(x,0.00282,-1)+b
```
