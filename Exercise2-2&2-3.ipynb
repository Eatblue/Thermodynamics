# 연습문제 2-2

세 부분계 $A_{1},A_{2},A_{3}$ 로 이루어진 결합계 $A_{0}$가 있다. 각 부분계의 자유도 수는 각각 $f_{1}=2, f_{2}=4, f_{3}=8$이고, 에너지는 각각 $E_{1}, E_{2}, E_{3}$이다. 결합계의 총 에너지가 $E_{0}$일 떄, 결합계의 모든 에너지분포에 대한 상태수를 구하고, 가장 가능한 에너지분포의 상태수가 차지하는 비율을 구하여라

python의 data analysis package인 pandas를 이용해서 표를 그려보자!

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
```

```python
f1,f2,f3=2,4,8
data={'Degree of freedom': [f1,f2,f3]}
frame=DataFrame(data,index=['A1','A2','A3'])# 간단한 데이터 프레임
frame
```

$\Omega(E)=E^{f/2}$ 라고 가정하고 문제를 풀자. 이제 for loop를 이용해서 $(E_1,E_2,E_3,\Omega_1,\Omega_2,\Omega_3,\Omega_0)$ 세트들을 구하자

```python
E0=4
Emat=np.array([[0,0,0,0,0,0,0]]) #더미 세트: 더미 세트를 없앨 수 있는 방법은?

for i in range(E0+1):
    E1=i
    for j in range(E0-E1+1):
        E2=j
        E3=E0-(E1+E2)
        Omega1=E1**(f1/2)
        Omega2=E2**(f2/2)
        Omega3=E3**(f3/2)
        Omega0=Omega1*Omega2*Omega3
        Emat=np.append(Emat,[[E1,E2,E3,Omega1,Omega2,Omega3,Omega0]],axis=0)
        
Emat
```

pandas의 데이터프레임으로 구현해보자

```python
frame1=DataFrame(Emat,columns=['E1','E2','E3','Omega1','Omega2','Omega3','Omega0'])
frame1
```

Omega0 열들을 따로 빼서 비율을 계산한다.

```python
frame1.Omega0  # 이렇게 따로 빼면 독립적인 column처럼 취급할 수 있다.
```

```python
irange=int(len(frame1.Omega0)) # =16
irange
```

```python
tot=np.sum(frame1.Omega0)
percentage=[]

for i in range(irange):
    percentage.append(100*frame1.Omega0[i]/tot)
percentage
```

```python
frame1=DataFrame(Emat,percentage,columns=['E1','E2','E3','Omega1','Omega2','Omega3','Omega0'])
frame1
```

가장 가능한 에너지 분포의 상태수는 (1,1,2) 이고 비율은 72.727273% 이다. 따라서 이 상태로 결합될 확률은 72.727273%!

# 연습문제 2-3

한편, $f_{1}=2\times10^{24}, f_{2}=4\times10^{24}, f_{3}=8\times10^{24}$ 일 때, 위의 과정을 되풀이해보면,

```python
f1,f2,f3=2e24,4e24,8e24

E0=4
Emat2=np.array([[0,0,0,0,0,0,0]]) #더미 세트: 더미 세트를 없앨 수 있는 방법은?

for i in range(E0+1):
    E1=i
    for j in range(E0-E1+1):
        E2=j
        E3=E0-(E1+E2)
        Omega1=E1**(f1/2)
        Omega2=E2**(f2/2)
        Omega3=E3**(f3/2)
        Omega0=Omega1*Omega2*Omega3
        Emat2=np.append(Emat2,[[E1,E2,E3,Omega1,Omega2,Omega3,Omega0]],axis=0) #Emat에 항목들 추가하기
        
frame2=DataFrame(Emat2,columns=['E1','E2','E3','Omega1','Omega2','Omega3','Omega0']) #데이터프레임 정의하기

tot=np.sum(frame2.Omega0) #Omega0 element 들의 총 합
irange=len(frame2.Omega0) #Omega0 element의 갯수
percentage=[]  #percentage를 담을 empty list

for i in range(irange):
    percentage.append(100*frame2.Omega0[i]/tot)

frame2=DataFrame(Emat2,percentage,columns=['E1','E2','E3','Omega1','Omega2','Omega3','Omega0'])
```
